```python
# import modules 
from yahpo_gym import BenchmarkSet, local_config
from joblib import Parallel, delayed
import random
import pandas as pd
import matplotlib as plt
import utils_BO as bo

# yahpo configurations
local_config.init_config()
local_config.set_data_path('yahpo_data-main')
```

```python
# minimal example
bench = BenchmarkSet('rbv2_xgboost', '12')
space = bench.get_opt_space()
cfg = bo.sample_config('rbv2_xgboost', space)
res = bench.objective_function(cfg)
print(cfg)
print(res[0])
```

```python
# BO Benchmark XGBoost
random.seed(13112025)

model = 'rbv2_xgboost'
metric = 'acc'
n_outer = 20
task_ids = BenchmarkSet('rbv2_xgboost').instances

final_df = pd.DataFrame()

results = Parallel(n_jobs=-1, backend='loky')(
    delayed(bo.run_BO)(model, metric, task, n_outer)
    for task in task_ids
)

final_df = pd.concat(results, axis = 0, ignore_index=True)

final_df.to_csv('xgboost_bo.csv', index = False)
```
